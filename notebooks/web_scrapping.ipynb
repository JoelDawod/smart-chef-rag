```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}
```

```python
urls =[]
for i in range(1,17):
    page = requests.get(f'https://www.atyabtabkha.com/wp-sitemap-posts-recipe-{i}.xml')
    soup = BeautifulSoup(page.content, features="xml")
    loc_tags = soup.find_all('loc')
    for loc in loc_tags:
        url = loc.text
        urls.append(url)

    print(len(urls))
```

```python
for row in df.values[0:1]:
    page = requests.get(row[0])
    soup = BeautifulSoup(page.content)

    section1 = soup.find('section', {'class':'section-secondary'})
    div1 = section1.find('div', {'class':'section__content'})
    title = div1.find('h1').get_text()
    print("الأسم: " + title)

    section2 = soup.find('section', {'class':'section section--alt'})
    ul2 = section2.find('ul',{'class':'recipe-list-products'})
    li2_list = ul2.find_all('li')
    ingredients=''
    for i in li2_list:
        ingredients+=i.get_text()+'\n'
    print(ingredients)

    section3 = soup.find('section', {'class':'section section--alt'})
    ul3 = section3.find('ul',{'class':'list-recipe'})
    li3_list = ul3.find_all('li')
    steps = ''
    for i in li3_list:
        steps+= i.get_text()
    #steps = steps.replace('\n\n','\n').replace('\n\n','\n').replace('\n\n','\n').replace('\n\n','\n')

    steps = "\n".join([line for line in steps.splitlines() if line.strip()])
    print(steps)
```
