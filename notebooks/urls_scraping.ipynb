```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
urls =[]
for i in range(1,17):
    page = requests.get(f'https://www.atyabtabkha.com/wp-sitemap-posts-recipe-{i}.xml')
    soup = BeautifulSoup(page.content, features="xml")
    loc_tags = soup.find_all('loc')
    for loc in loc_tags:
        url = loc.text
        urls.append(url)

    print(len(urls))
```

```python
import pandas as pd

df = pd.DataFrame(urls, columns=['Recipe URL'])

# 2. Export the DataFrame directly to a CSV file
df.to_csv('atyabtabkha_recipes_urls.csv', index=False, encoding='utf-8')
```
